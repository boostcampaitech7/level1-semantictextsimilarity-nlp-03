```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')

bins = [round(x * 0.1, 1) for x in range(0, 51)]
data['label_bins'] = pd.cut(data['label'], bins=bins, include_lowest=True)

label_distribution = data['label_bins'].value_counts().sort_index()
label_distribution = label_distribution[label_distribution > 0]

def format_label(x):
    return '0.0' if x.right == 0.1 else str(x.right)

formatted_labels = label_distribution.index.map(format_label)

output_text = '\n'.join([f"{label}: {count}" for label, count in zip(formatted_labels, label_distribution.values)])
print(output_text)

plt.figure(figsize=(10,6))
plt.bar(formatted_labels, label_distribution.values, width=0.4)
plt.xlabel('Label Ranges (Similarity Score)')
plt.ylabel('Frequency')
plt.title('Label Distribution (basic data & label)')
plt.xticks(rotation=45) 
plt.show()
```
